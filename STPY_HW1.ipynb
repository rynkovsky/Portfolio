### Задание 1. Загрузка данных

```python
import pandas as pd
import numpy as np
```

```python
names = [1, 2, 4, 5, 6, 7, 11, 23]
horses_data = pd.read_csv('E:\\Python\\STPY-26\\STPY_HW1\\horse_data.csv', sep = ',',  na_values = '?')

df = horses_data.iloc[:, names]
df.columns = ['surgery', 'Age', 'rectal temperature','pulse','respiratory rate','temperature of extremities','pain','outcome',]
df.head()
```

```python
df.info()
```

### Задание 2. Первичное изучение данных

```python
df.describe(include='all')
```

```python
print('Значение моды: ', df.pulse.round().mode()[0])
```

```python
q1 = df.pulse.quantile(0.25)
q3 = df.pulse.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df.pulse.between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
remove_outliers
```

```python
q1 = df.pain.quantile(0.25)
q3 = df.pain.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr) 
remove_outliers = df[df.pain.between(lower_bound, upper_bound, inclusive=True)].sort_values('pain')
remove_outliers 
```

```python
q1 = df.outcome.quantile(0.25)
q3 = df.outcome.quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df.outcome.between(lower_bound, upper_bound, inclusive=True)].sort_values('outcome')
remove_outliers
```

### Задание 3. Работа с пропусками

```python
# пропуски
print(df.isna().sum())
```

```python
# Новый датафрейм
new_df = df.copy()
```

#### Замена медианой

```python
new_df['pulse'] = df['pulse'].fillna(df['pulse'].median())
new_df['rectal temperature'] = df['rectal temperature'].fillna(df['rectal temperature'].median())
new_df['respiratory rate'] =  df['respiratory rate'].fillna(df['respiratory rate'].median())
```

#### Замена модой

```python
new_df['pain'] = df['pain'].fillna(df['pain'].mode()[0])
new_df['temperature of extremities'] = df['temperature of extremities'].fillna(df['temperature of extremities'].mode()[0])
```
