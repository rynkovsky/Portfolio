Сравним для начала значения некоторых столбцов для первых десяти стран из списка на диаграммах ниже.

```python
fig,ax = plt.subplots(figsize=(15,8))
plt.bar(df_2015.head(10)['Country'],df_2015.head(10)['Generosity'])
plt.ylabel("Genoriosity", fontsize=20,color="Green")
plt.xlabel("Country", fontsize=20,color="Blue")
```

```python
fig,ax = plt.subplots(figsize=(15,8))
plt.bar(df_2015.head(10)['Country'],df_2015.head(10)['Freedom'])
plt.ylabel("Freedom", fontsize=20,color="Green")
plt.xlabel("Country", fontsize=20,color="Blue")
```

```python
fig,ax = plt.subplots(figsize=(15,5))
plt.bar(df_2015.head(10)['Country'],df_2015.head(10)['Family'])
plt.ylabel("Family", fontsize=20,color="Green")
plt.xlabel("Country", fontsize=20,color="Blue")
```

Построим тепловую карту, для этого создадим матрицу корреляции.

```python
df_corr = df_2015.corr()
df_corr
```

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.heatmap(df_corr)
```

Из тепловой карты видно, что есть сильная прямая зависимость между значениями ВВП и Happiness Score, чуть меньше, но тоже ярко выражена зависимость Happiness Score от значений здоровья и Семьи. 
А между Happiness Score и Happiness Rank наоборот, зависимость обратная.
Это же подтверждается ниже на графиках.

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 7))
sns.scatterplot(x = df_2015["Health (Life Expectancy)"], y = df_2015["Happiness Score"], size = df_2015["Economy (GDP per Capita)"])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 7))
sns.scatterplot(x = df_2015["Family"], y = df_2015["Happiness Score"], size = df_2015["Economy (GDP per Capita)"])
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 7))
sns.scatterplot(x = df_2015["Happiness Rank"], y = df_2015["Happiness Score"], size = df_2015["Economy (GDP per Capita)"])
```

Построим боксплоты для того, чтобы посмотреть выбросы.

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 7))
sns.boxplot(data = df_2015.drop(['Happiness Rank', 'Happiness Score', 'Standard Error', 'Dystopia Residual'], axis=1), ax=ax)
plt.show()
```
