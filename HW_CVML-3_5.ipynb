```python
import numpy as np
import pandas as pd 
```

```python
train = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
test = pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=16, svd_solver='randomized',
          whiten=True)
```

```python
X, y = train.drop('label', axis=1), train['label']
```

```python
X_pca = pca.fit_transform(X)
```

```python
X_pca.shape
```

```python
from sklearn import svm
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import train_test_split
y_train, y_val, x_train, x_val = train_test_split(
    y, X_pca, test_size=0.2,
    random_state=42)
```

```python
model = svm.SVC(probability=True)
```

```python
model.fit( x_train, y_train ) 
```

```python
model.score( x_val, y_val ) 
```

```python
X_t = test
X_t_pca = pca.transform(X_t)
```

```python
# Use the model to make predictions
predicted_label = model.predict(X_t_pca)
# We will look at the predicted prices to ensure we have something sensible.
print(predicted_label)
```

```python
with open('submit.txt', 'w') as dst:
    dst.write('ImageId,Label\n')
    for i, p in enumerate(predicted_label, 1):
        dst.write('%s,%s\n' % (i, p))
```
